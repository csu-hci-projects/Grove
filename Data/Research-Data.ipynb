Plot of level score per participant.

```python
# 8 columns / 9 rows

counterbalance_data = [
        [1, '1/2/3', 3, 501, 8, 395, 2, 253],
        [2, '2/3/1', 4, 454, 2, 261, 3, 171],
        [3, '3/1/2', 3, 549, 12, 465, 11, 382],
        [4, '3/2/1', 8, 661, 6, 600, 6, 236],
        [5, '2/1/3', 4, 365, 7, 294,  4, 245],
        [6, '1/3/2', 9, 421, 9, 241, 20, 280],
        [7, '1/2/3', 4, 283, 9, 295, 9, 246],
        [8, '2/3/1', 10, 547, 9, 307, 6, 217],
        [9, '1/3/2', 8, 366, 8, 297, 10, 292],
       ]
 
counterbalance_df = pd.DataFrame(counterbalance_data, columns = ['Participant', 'Counterbalance',
                                   'First Level Err.', 'First Level Time', 'Second Level Err.', 'Second Level Time',
                                   'Third Level Err.', 'Third Level Time'])
counterbalance_df
```

Plot of level times per participant.

```python
Error = counterbalance_df[['First Level Err.', 'Second Level Err.', 'Third Level Err.']]

print("First Level Err.:")
print(np.mean(Error['First Level Err.']))
print(np.std(Error['First Level Err.']))

print("Second Level Err.:")
print(np.mean(Error['Second Level Err.']))
print(np.std(Error['Second Level Err.']))

print("Third Level Err.:")
print(np.mean(Error['Third Level Err.']))
print(np.std(Error['Third Level Err.']))

Error.plot(kind='bar')
plt.legend()
plt.title('Level Errors per Participant')
plt.xlabel('Participant')
plt.ylabel('Number of Errors')
```

```python
Time = performance_df[['L1 Time', 'L2 Time', 'L3 Time']]

print("Level 1 Times:")
print(np.mean(Time['L1 Time']))
print(np.std(Time['L1 Time']))

print("Level 2 Times:")
print(np.mean(Time['L2 Time']))
print(np.std(Time['L2 Time']))

print("Level 3 Times:")
print(np.mean(Time['L3 Time']))
print(np.std(Time['L3 Time']))

Time.plot()
plt.legend()
plt.title('Level Times per Participant')
plt.xlabel('Participant')
plt.ylabel('Level Time')
```
